# 1. Conectarse al enlace

```python
# Libraries
import requests
import json
import pandas as pd

pd.set_option('max_columns', None)
```

### Crear una función para obtener los datos

```python
def get_data(url):
  response = requests.get(url)
  results = response.json()
  return results
```

```python
# Call the function
url = 'https://api.stackexchange.com/2.2/search?order=desc&sort=activity&intitle=perl&site=stackoverflow'
data = get_data(url)
```

```python
# Inspect the data
data
```

### Guardar los datos en un dataframe

```python
# JSON a DataFrame
df = pd.json_normalize(data['items'])
df.head(2)
```

```python
# Dimension of the dataframe
print('Total Columns: {}'.format(df.shape[1]))
print('Total rows:    {}'.format(df.shape[0]))
```

```python
# Missing Values
df.isna().sum()
```

```python
print('Respuesta más vieja:  {}'.format(df['creation_date'].min()))
print('Respuesta más actual: {}'.format(df['creation_date'].max()))
```
